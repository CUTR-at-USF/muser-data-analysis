##### Extract the event date and time

```
def extract_date_from_mills(time_stamp):
  seconds = time_stamp/1e3
  dt = datetime.datetime.fromtimestamp(seconds)
  return dt.strftime('%Y-%m-%d')

def extract_time_from_mills(time_stamp):
  seconds = time_stamp/1e3
  dt = datetime.datetime.fromtimestamp(seconds)
  return dt.strftime('%I:%M:%S %p')
```

```
extract_time_from_mills(1.605021e+12)
```

```
muser_data["event_trigger_date"] = muser_data["event_current_time_in_milliseconds"].apply(extract_date_from_mills)
muser_data["event_trigger_time"] = muser_data["event_current_time_in_milliseconds"].apply(extract_time_from_mills)
```

##### User daily app usage

```
user_behavior_data.columns
```

```
user_behavior_data.dtypes
```

```
unique_behavior_data = user_behavior_data.drop_duplicates(['user_id', 'song_id', 'event_type', 'event_trigger_date', 'event_trigger_time']).sort_values(["event_trigger_date", "event_trigger_time"]).reset_index().drop(["index"], axis=1)
```

```
unique_behavior_data.head(20)
```

```
unique_behavior_data["duration"] = 0
```

```
for idx in range(1, len(unique_behavior_data)):
  user = unique_behavior_data.loc[idx, "user_id"]
  song = unique_behavior_data.loc[idx, "song_id"]
  event = unique_behavior_data.loc[idx, "event_type"]
  trigger_time = unique_behavior_data.loc[idx, "event_current_time_in_milliseconds"]
  duration = 0
  if user != unique_behavior_data.loc[idx-1, "user_id"]:
    continue
  elif song != unique_behavior_data.loc[idx-1, "song_id"]:
    continue
  else:
    if event == "PAUSE" and unique_behavior_data.loc[idx-1, "event_type"] == "PLAY":
      duration = trigger_time - unique_behavior_data.loc[idx-1, "event_current_time_in_milliseconds"]
      unique_behavior_data.loc[idx, "duration"] = duration/(1000*60)
```

```
unique_behavior_data["duration"].max()
```

```
unique_behavior_data[unique_behavior_data["duration"]==unique_behavior_data["duration"].max()]
```

```
app_usage_data = pd.DataFrame(unique_behavior_data.groupby(["user_id", "event_trigger_date"])["duration"].sum()).reset_index()
app_usage_data.head()
```

```
fig = px.line(app_usage_data, x="event_trigger_date", y="duration", color="user_id", title="Daily app usage by users")
fig.show()
```

##### Weekly user behavior pattern

```
user_behavior_data.dtypes
```

```
user_behavior_data.dtypes
```

```
def extract_week_of_month(date):
  day = int(date.split("-")[2])
  return (day - 1)//7 + 1
```

```
user_behavior_data["week_of_month"] = user_behavior_data["event_trigger_date"].apply(extract_week_of_month)
user_behavior_data.head()
```

```
user_behavior_data.tail()
```

##### Weekly app usage

```
weekly_user_behavior = pd.DataFrame(user_behavior_data["week_of_month"].value_counts()).reset_index()
weekly_user_behavior.columns = ["week_of_month", "count"]
```

```
weekly_user_behavior
```

```
fig = px.bar(weekly_user_behavior, x="week_of_month", y="count", title="Weekly app usage")
fig.show()
```

##### Weekly user behavior

```
user_weekly_data = pd.DataFrame(data=user_behavior_data.groupby("user_id")["week_of_month"].value_counts().values, index=user_behavior_data.groupby("user_id")["week_of_month"].value_counts().index).reset_index()
user_weekly_data.columns = ["user_id", "week_of_month", "count"]
user_weekly_data["week_of_month"] = user_weekly_data["week_of_month"].astype("category")
user_weekly_data
```

```
user_weekly_data = user_weekly_data[user_weekly_data["count"]!=user_weekly_data["count"].max()]
user_weekly_data.sort_values("count")
```

```
fig = px.bar(user_weekly_data, x="week_of_month", y="count", color="user_id", title="Weekly user behavior")
fig.show()
```
